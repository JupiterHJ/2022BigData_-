#3. Grouping

```python
import pandas as pd
df4 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')

#45. 각 host_name의 빈도수 정렬
df4.groupby('host_name').size()

#★46. host_name의 빈도수를 구한 뒤, 빈도수 칼럼이 'counts'이고, 내림차순 정렬한 데이터프레임 만들기.
df4.groupby('host_name').size().to_frame().rename(columns={0:'counts'}).sort_values('counts',ascending=True)

#★48. 'neighbourhood_group의 값에 따른 neighbourhood'의 컬럼값 개수 중, 'neighnourhood_group'의 각 항목 최댓값
df4.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().groupby(['neighbourhood_group'], as_index=False).max()
```

```python
#★49. 'neighbourhood_group에서 price 값'의 평균,분산,최대,최소 값
df4.groupby(['neighbourhood_group']).price.agg(['mean','var','max','min']).unstack()

#★52. neighbourhood 값과 neighbourhood_group 값에 따른, price의 평균을, indexing 없이 구하기
df4.groupby(['neighbourhood','neighbourhood_group']).price.mean().unstack()

#★55. 'neighbourhood_group 값에 따른 room_type' 컬럼 숫자를 구하고, 'neighbourhood_group'을 기준으로 각 값의 비율
Ans = df4.groupby(['neighbourhood_group','room_type']).size().unstack()
Ans.loc[:,:] = (Ans.values /Ans.values.sum(axis=1).reshape(-1,1))
Ans
```

#4. Apply, Map

```python
df5=pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv',index_col=0)

#★57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑

dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   }

df5['newIncome'] = df5.Income_Category.map(lambda x: dic[x])
```

```python
#★58. Income_Category의 카테고리를 apply 함수를 이용하여 newIncome 컬럼에 매핑

def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'

df5['newIncome']  =df5.Income_Category.apply(changeCategory)
```

```python
#59. Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의 (0~9 : 0 , 10~19 :10 , 20~29 :20 …)
df5['AgeState']  = df5.Customer_Age.map(lambda x: x//10 *10)

#60. Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의
df5['newEduLevel'] = df5.Education_Level.map(lambda x: 1 if 'Graduate' in x else 0)

#62. Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼
def check(x):
    if x.Marital_Status =='Married' and x.Card_Category =='Platinum':
        return 1
    else:
        return 0

df5['newState'] = df5.apply(check,axis=1)

#64. Gender 컬럼값 M인 경우 male F인 경우 female로 값을 변경하여 Gender 컬럼에 새롭게 정의
def changeGender(x):
    if x =='M':
        return 'male'
    else:
        return 'female'
df5['Gender'] = df5.Gender.apply(changeGender)
```

#5. Time Series

```python
import pandas as pd

df6 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv')
df7 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')

#★76. 년-월-일:시 (2021-05-15:15) 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경
def change_date(x):
    import datetime
    date = x.split(":")[0]
    hour = x.split(':')[1]
    
    if hour =='24':
        hour ='00:00:00'      
        FinalDate = pd.to_datetime(date +" "+hour)+datetime.timedelta(days=1)
        
    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date +" "+hour)
    
    return FinalDate

df7['(년-월-일:시)'] = df7['(년-월-일:시)'].apply(change_date)
df7
```

```python
#81.날짜 컬럼을 index로 만들기
a=df7.set_index('(년-월-일:시'), inplace=True, drop=True)
```

```python
#★65. Yr_Mo_Dy을 판다스에서 인식할 수 있는 datetime64타입으로 변경
df6.Yr_Mo_Dy = pd.to_datetime(df6.Yr_Mo_Dy)

#★66.Yr_Mo_Dy에 존재하는 년도의 유일값을 모두 출력
df6.Yr_Mo_Dy.dt.year.unique()

#69. weekday컬럼을 만들고 요일별로 매핑
df6['weekday'] = df6.Yr_Mo_Dy.dt.weekday

#77. 일자별 영어요일 이름을 dayName 컬럼에 저장
df7['dayName']  =df7['(년-월-일:시)'].dt.day_name()

#★78. 일자별 각 PM10등급의 빈도수를 파악
df7.groupby(['dayName','PM10등급'],as_index=False).size().pivot(index='dayName',columns='PM10등급',values='size').fillna(0)
```

```python
#★68. 년도별 각컬럼의 평균값
df6.groupby(df6.Yr_Mo_Dy.dt.year).mean()

#★73.년도 - 월을 기준으로 모든 컬럼의 평균값
df6.groupby(df6.Yr_Mo_Dy.dt.to_period('M')).mean()

#RPT와 VAL의 컬럼을 일주일 간격으로 각각 이동평균한값
df6[['RPT','VAL']].rolling(7).mean()
```

```python
#★72. 모든 결측치는 컬럼기준 직전의 값으로 대체, 첫번째 행에 결측치가 있을경우 뒤에 있는 값으로 대체
df6 = df6.fillna(method='ffill').fillna(method='bfill')

#★74. RPT 컬럼의 값을 일자별 기준으로 1차차분
df6['RPT'].diff().unique()

#★79. 시간이 연속적으로 존재하며 결측치가 없는지 확인
# 시간을 차분했을 경우 첫 값은 nan, 이후 모든 차분값이 동일하면 연속이라 판단한다.
check = len(df7['(년-월-일:시)'].diff().unique())
if check ==2:
    Ans = True
else:
    Ans = False
Ans
```

```python
#70.weekday컬럼을 기준으로 주말이면 1 평일이면 0의 값을 가지는 WeekCheck 컬럼
df6['WeekCheck']  = df6['weekday'].map(lambda x : 1 if x in [5,6] else 0)


#★67.Yr_Mo_Dy에 년도가 2061년 이상의 경우, 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의.
def fix(x):
  import datetime
  if x.year >= 2061:
    year = x.year - 100
  else:
    year = x.year
  return pd.to_datetime(datetime.date(year, x.month, x.day))
df6['Yr_Mo_Dy'] = df6['Yr_Mo_Dy'].apply(fix)


#★80. 오전 10시와 오후 10시(22시)의 PM10의 평균값
df7.groupby(df7['(년-월-일:시)'].dt.hour).mean().iloc[[10,22],[0]])
```
